```python
page = requests.get(url).text
soup = BeautifulSoup(page)
soup
```

```python
header1 = soup.find("h1", attrs={"class":"ssrcss-15xko80-StyledHeading e1fj1fc10"}).text
header1
```

```python
#header2 = soup.find_all("h2",attrs={"class":"ssrcss-y2fd7s-StyledHeading e1fj1fc10"}).text
header2 = soup.find_all('h2')
h2_texts = [line.text for line in header2]
h2_texts
```

```python
paragraphs = soup.find_all('p')
p_texts = [line.text for line in paragraphs]
#p_texts
article = " ".join(p_texts)
article
```

```python
# 1b
# Create a table that contains the following informations from the article 
#   using Regular expressions and Pandas
#     Father of the nation
#     Date of birth of "Father of the nation"
#     The kingmaker
#     The disillusioned queen
#     The dead prince
#     The 'royal rebels'
#     The mouthpiece
```

```python
import re
import pandas as pd
h2_table = []

DoB = re.search('\d+\s\w+\s\d+',article)

for item in h2_texts:
    h2_dict = {}
    
    # ADD THE MAIN TITLE OF THE h2 TO THE DICTIONARY
    main = re.search('\w.*\:',item)
    if main is not None:
        main1 = re.search('.*',main.group()) 
    else:
        main1 = None
    if main1 is not None:
        mn = main1.group()
    else:
        mn = None
    h2_dict['Title'] = mn
    
    cnt = re.search('\:\s.*',item)
    if cnt is not None:
        cnt1 = re.search('\s.*',cnt.group()) 
    else:
        cnt1 = None
    if cnt1 is not None:
        content = cnt1.group()
    else:
        content = None
    h2_dict['Content'] = content
    
    h2_table.append(h2_dict)
    
#h2_table
h2_df = pd.DataFrame(h2_table)

new_row = {'Title': 'Date of birth of the Father of the nation:', 'Content': DoB.group()}
h2_df = pd.concat([h2_df.iloc[:1], pd.DataFrame([new_row]), h2_df.iloc[1:]], ignore_index=True)

h2_df
```

# Question 2: Access log

```python
# An access_log.txt file is provided. Your mission, should you accept is to...
# Apply your knowledge of regular expressions and pandas to store all 
#    "IP Address", "Dates", "The requests" in a dataframe
```

```python
# --THESE ARE NOTES FOR MY EASY ACCESS--

# re.search(pattern, string) returns the first occurance of whatever you are looking for
# re.findall(pattern, string) extracts all the occurences of whatever you are looking for

# \w: matches all alphanumeic characters a-z, A-Z, 0-9 and _ and -.
# \d: matches digits, which means (0-9)
# \s: matches whitespace characters which means (tab, ) 
# \S: matches non-whitespace characters 
# \W: matches a single character from (a-z, A-Z, 0-9)
# \d: matches a single character digit from (0-9)

# \D: matches non-digits

# (.)fullstop: matches any character except the new line
# *: matches zero or more occurence
# +: matches one or more occurence
# ?: matches zero or one occurence

# ---FURTHER NOTES---
# \d{2,}: This matches any number with 2 or more digits
# \.: This matches the period in-between the digits in the ip address
# \d{2}: This matches any number with exactly 2 digits
# \/: This matches the forward slash in-between the date
```

```python
import re
import pandas as pd

txt_log = []
txt = open('access_log.txt', encoding="utf8").read() #'r' symbolises read mode

content = re.split('\"\-\"\s\"\-\"\s',txt)
content.pop(0)
#content

for item in content:
    logs_dict = {} # create dictionary for data storage
    
    # ADD DATE OF ACCESS TO DICTIONARY
    date = re.search('\d{2}\/\D{3}\/\d{4}',item)
    #print(date.group())
    if date is not None:
        DoAccess = re.search('.*',date.group()) # DoAccess means date of access
    else:
        DoAccess = None
    if DoAccess is not None:
        dateAccessed = DoAccess.group()
    else:
        dateAccessed = None
    logs_dict['Date Accessed'] = dateAccessed
    #print(logs_dict)

    # ADD IP ADDRESSES TO DICTIONARY
    ip = re.search('\d{2,}\.\d{3}\.\d{3}\.\d{2,}',item)
    #print(date.group())
    if ip is not None:
        ip1 = re.search('.*',ip.group()) 
    else:
        ip1 = None
    if ip1 is not None:
        ipAdd = ip1.group()
    else:
        ipAdd = None
    logs_dict['IP Addresses'] = ipAdd
    #print(logs_dict)
    
    # ADD REQUEST TO DICTIONARY
    req = re.search('\"\D{3,}\w\S.\d{1}\.\d{1}\"',item)
    #print(req.group())
    if req is not None:
        req1 = re.search('.*',req.group())
    else:
        req1 = None
    if req1 is not None:
        txtRequest = req1.group()
    else:
        txtRequest = None
    logs_dict['The Request'] = txtRequest
    print(logs_dict)
    txt_log.append(logs_dict)
```

```python
log_df = pd.DataFrame(txt_log)
log_df.to_csv('access_log.csv', index = True) #exports the created data frame as a .csv file
log_df
```
